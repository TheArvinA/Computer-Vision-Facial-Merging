# Dataset Organizer

This is the python script used to organize the dataset found [here](https://www.kaggle.com/datasets/atulanandjha/lfwpeople)

```python
import os
import shutil

#@markdown Location of the image dataset
source_dir = "./dataset" #@param {type:"string"}

#@markdown Location of where to store all the images
dest_dir = "./dataset_images" #@param {type:"string"}

if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)

```

```python
# Walk through the source directory
for foldername, _, filenames in os.walk(source_dir):
    for filename in filenames:
        # Construct the full source path of the image
        source_path = os.path.join(foldername, filename)

        # Construct the full destination path
        destination_path = os.path.join(dest_dir, filename)

        # If a file with the same name exists, modify the name to avoid overwrite
        if os.path.exists(destination_path):
            base, ext = os.path.splitext(filename)
            counter = 1
            while os.path.exists(destination_path):
                destination_path = os.path.join(dest_dir, f"{base}_{counter}{ext}")
                counter += 1

        # Move the image to the destination folder
        shutil.move(source_path, destination_path)
        print(f"Moved: {source_path} -> {destination_path}")
```
